## BigQuery on Jupyter

```python
%load_ext google.cloud.bigquery
```

```python
%%bigquery
SELECT *
FROM `hololive-youtube-analysis.hololiveyoutube.hololive_channel`
```

```python
%%bigquery data
SELECT 
ch.full_name, 
ch.branch, 
ch.gen, 
vid.title, 
vid.published_date, 
vid.duration, 
vid.views, 
vid.likes
FROM hololiveyoutube.hololive_videos AS vid
INNER JOIN hololiveyoutube.hololive_channel AS ch
ON ch.channel_id = vid.channel_id;
```

### Analysis to Moona Hosinova

```python
talent = 'Moona Hoshinova'
moona = data.loc[data.full_name == talent]
```

```python
moona['gen'].unique()
```

```python
import seaborn as sns
```

#### Views over time

```python
sns.set(rc={'figure.figsize':(12,8)})
sns.lineplot(
    data=moona,
    x = 'published_date',
    y = 'views'
)
```

#### Views-Like relation

```python
sns.scatterplot(
    data=moona,
    x = 'views',
    y = 'likes'
)
```

```python
import pandas as pd
moona['duration'] = pd.to_timedelta(moona['duration'])
```

```python
moona.dtypes
```

```python
moona['duration'] = moona['duration'].astype('timedelta64[s]')
```

#### Stream duration (video > 10min) to views

```python
sns.scatterplot(
    data=moona.loc[moona['duration'] > 600],
    x = 'duration',
    y = 'views'
)
```

```python
sns.scatterplot(
    data=moona.loc[moona['duration'] > 600],
    x = 'duration',
    y = 'likes'
)
```
